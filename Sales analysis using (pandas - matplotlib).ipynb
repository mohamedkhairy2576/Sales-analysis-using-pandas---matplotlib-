```python
df = pd.read_csv(r"D:\pandas_data_test\Orders.csv" , skiprows  = 4 ) 
# Skip 4 rows for column headers
df.head()
# show first rows of data
```

```python
df.dropna(how = 'all' , inplace = True)
# Delete empty row when it is completely empty
df.isna().sum()
```

```python
df.dtypes
# show data type for each column
```

```python
df['Order ID']   = df['Order ID'].astype(int)
df['Order Date'] =  pd.to_datetime(df['Order Date'] , dayfirst = True)
df['Quantity']   = df['Quantity'].astype(int)
# edit 3 columns to new data type
```

```python
df.dtypes
```

```python
df.describe()
# Statistical information about the data
```

```python
df.sample(4)
```

```python
df.set_index('Order ID' , inplace = True)
#Place the order column as an index to speed up performance and facilitate searching
```

```python
df.tail(4)
```

```python
df.insert(7 ,'Customer Full Name' ,  df['Customer First Name']+' '+df['Customer Last Name'] )
df['Customer Full Name']
# insert cu full name
```

```python
df.drop(columns = ['Customer First Name' , 'Customer Last Name'] , inplace = True)
# and drop cu first name and last name
```

```python
df['Email']
```

```python
df['Email'] = df['Email'].str.lower()
# Convert email column to lower
df['Email']
```

```python
df['Discount'].fillna(0 , inplace = True )
# replace any null to 0 , For mathematical operations
```

```python
df['Total Sales'] = df['Quantity']    * df['Unit Price']
df['Profit']      = df['Total Sales'] - df['Discount']
df[['Total Sales' , 'Profit']]
# add total_sales and profit columns
```

```python
top_custmers = df.groupby(['Customer Full Name','Email'])['Profit'].sum().sort_values(ascending = False)
top_custmers.head(10)
#Highest Profiting Customers
```

```python
top_sub_category = df.groupby('Sub Category')['Profit'].sum().sort_values(ascending = False)
top_sub_category = top_sub_category.head(10)
```

```python

plt.figure(figsize=(12, 7))
wedges, texts, autotexts = plt.pie(
    top_sub_category.values, 
    labels=top_sub_category.index,
    shadow=True,  
    autopct='%1.1f%%',  
    textprops={'fontsize': 10, 'color': 'black'}, 
    startangle=90  
)

plt.legend(top_sub_category.index, title="Sub-Categories", loc="best")

for autotext in autotexts:
    autotext.set_color('white')  
    autotext.set_fontweight('bold') 
plt.title('Top Sub-Categories by Total Sales', fontsize=15, color='b')

plt.axis('equal')

plt.show()
```

```python
sales_cities = df.groupby('City')['Total Sales'].sum().sort_values(ascending = False)
```

```python
plt.figure(figsize = (12,6))
plt.bar(sales_cities.index , sales_cities.values , color = 'skyblue'  )
plt.xticks(rotation = 75)
plt.title('Total Sales For Each City' , fontsize = 15 , color = 'black'  )
plt.ylabel('Total Sales' )
for ind , v in enumerate(sales_cities.values) :
    plt.text(ind, v ,v, ha = 'center' , va = 'bottom'  , rotation = 60 , color = 'black' )
plt.tight_layout()
plt.show()
```

```python
sales_by_categories = df.groupby('Category')['Total Sales'].sum().sort_values(ascending = False)
```

```python
plt.figure(figsize = (12,6))
plt.bar(sales_by_categories.index , sales_by_categories.values , color = 'b'   , align = 'edge' , hatch = '+')
plt.xticks(rotation = 75)
for ind , v in enumerate(sales_by_categories.values) :
    plt.text(ind,v,v   , ha = 'left' , va = 'bottom' , c = 'black' , fontsize = 10)
plt.ylabel('Total Sales')
plt.title('Total Sales By Categories' , fontsize = 15 , y = 1.09)
plt.tight_layout()
plt.show()
```
